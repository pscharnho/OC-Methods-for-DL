# Convolutional Comparison

ConvNets with MSA and backpropagation training are compared. The MNIST dataset for image recognition is used and splitted into a training set with 60000 samples and a test set with 10000 samples. 
Each net has 3 convolutional layers and 2 fully connected layers. After two subsamplings there are 12 channels with size 7x7 for each. This gives a flattened input of 12*7*7 = 588 for the fc layers.
The nets are trained for 10 epochs and the results are averaged over 10 iterations of training.

```python
from Dataset.Dataset import loadMNIST

train_loader, test_loader = loadMNIST('Dataset/input/mnist_train.csv', 'Dataset/input/mnist_test.csv')

test_set_size = 10000
```

## ConvNet

```python
from Networks.ResNet import ConvNet
import numpy as np

num_epochs = 10

basic_save_key = 'Convolutional_results//ConvNet//ConvNet'

train_results_per_epoch = np.zeros(num_epochs)
test_results_per_epoch = np.zeros(num_epochs)

for i in range(10):
    net = ConvNet(num_conv=3, num_channels=[1,3,6,12], subsample_points=[0,1], num_fc=2,sizes_fc=[588,900,10])
    net.set_test_tracking(True)
    net.train(num_epochs,train_loader, testloader=test_loader)
    train_results_per_epoch += net.avg_correct_pred.numpy()
    test_results_per_epoch += net.test_results.numpy()
    
    train_save_key = basic_save_key+'train_'+str(i)+'.csv'
    test_save_key = basic_save_key+'test_'+str(i)+'.csv'
    np.savetxt(train_save_key, net.avg_correct_pred.numpy(), delimiter=',')
    np.savetxt(test_save_key, net.test_results.numpy(), delimiter=',')
    
train_results_per_epoch = train_results_per_epoch / 10
test_results_per_epoch = test_results_per_epoch / 10
train_save_key = basic_save_key+'train.csv'
test_save_key = basic_save_key+'test.csv'
np.savetxt(train_save_key, train_results_per_epoch, delimiter=',')
np.savetxt(test_save_key, test_results_per_epoch, delimiter=',')
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
epochs = np.arange(1,num_epochs+1,1)
ax.plot(epochs, train_results_per_epoch, label='train')
ax.plot(epochs, test_results_per_epoch/test_set_size, label='test')
ax.set_ylim(0,1.1)
ax.set_xlabel('Epoch')
ax.set_ylabel('Average Accuracy')
ax.legend(loc='best')
```

## MSAConvNet

```python
from Networks.ResNet import MSAConvNet

num_epochs = 10

basic_save_key = 'Convolutional_results//MSAConvNet//MSAConvNet'

train_results_per_epoch = np.zeros(num_epochs)
test_results_per_epoch = np.zeros(num_epochs)

for i in range(10):
    net = MSAConvNet(num_conv=3, num_channels=[1,3,6,12], subsample_points=[0,1], num_fc=2,sizes_fc=[588,900,10], batchnorm=True)
    net.set_test_tracking(True)
    net.train(num_epochs,train_loader, testloader=test_loader)
    train_results_per_epoch += net.avg_correct_pred.numpy()
    test_results_per_epoch += net.test_results.numpy()
    
    train_save_key = basic_save_key+'train_'+str(i)+'.csv'
    test_save_key = basic_save_key+'test_'+str(i)+'.csv'
    np.savetxt(train_save_key, net.avg_correct_pred.numpy(), delimiter=',')
    np.savetxt(test_save_key, net.test_results.numpy(), delimiter=',')
    
train_results_per_epoch = train_results_per_epoch / 10
test_results_per_epoch = test_results_per_epoch / 10
train_save_key = basic_save_key+'train.csv'
test_save_key = basic_save_key+'test.csv'
np.savetxt(train_save_key, train_results_per_epoch, delimiter=',')
np.savetxt(test_save_key, test_results_per_epoch, delimiter=',')
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
epochs = np.arange(1,num_epochs+1,1)
ax.plot(epochs, train_results_per_epoch, label='train')
ax.plot(epochs, test_results_per_epoch/test_set_size, label='test')
ax.set_ylim(0,1.1)
ax.set_xlabel('Epoch')
ax.set_ylabel('Average Accuracy')
ax.legend(loc='best')
```
